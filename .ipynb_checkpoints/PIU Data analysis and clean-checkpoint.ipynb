# BUSINESS PROBLEM

The Tanzanian Ministry of Water (TMOW) is having trouble keeping track of all the water pumps that are in operation around the country. They would like a way of predicting which wells are non-functioning or which require some maintenance. 

The aim of this project is to help the TMOW assign predictions to all its wells in the form of the following categories: 

- The well is functional
- The well is functional but needs repair
- The well is non functional

```python
def opt_plots(opt_model):
    
    """
    This function plots our parameter heat map to visualise where our model perfoms best!
    """
    opt = pd.DataFrame(opt_model.cv_results_)
    cols = [col for col in opt.columns if ('mean' in col or 'std' in col) and 'time' not in col]
    params = pd.DataFrame(list(opt.params))
    opt = pd.concat([params,opt[cols]],axis=1,sort=False)
    
    plt.figure(figsize=[15,4])
    plt.subplot(121)
    sns.heatmap(pd.pivot_table(opt,index='max_depth',columns='min_samples_leaf',values='mean_train_score')*100)
    plt.title('ROC_AUC - Training')
    plt.subplot(122)
    sns.heatmap(pd.pivot_table(opt,index='max_depth',columns='min_samples_leaf',values='mean_test_score')*100)
    plt.title('ROC_AUC - Validation')
    return opt
```

```python
from sklearn.metrics import roc_auc_score
def scores(model,X_train,X_val,y_train,y_val, test=True):
    
    """
    This function generates a score based on the probability of our model identifying the correct class.
    """
    train_prob = model.predict_proba(X_train)[:,1]
    val_prob = model.predict_proba(X_val)[:,1]
    train = roc_auc_score(y_train,train_prob)
    val = roc_auc_score(y_val,val_prob)
    if test:
        print('train:',round(train,2),'validation:',round(val,2)) #validation was 'test'
    else:
        print('train:',round(train,2),'test:',round(val,2))
        
```

```python
def annot(fpr,tpr,thr):
    """
    This function is called in the roc_plot function and serves to annotate the roc curves.
    """
    k=0
    for i,j in zip(fpr,tpr):
        if k %50 == 0:
            plt.annotate(round(thr[k],2),xy=(i,j), textcoords='data')
        k+=1
```

```python
def roc_plot(model,X_train,y_train,X_val,y_val, test=True): 
    
    """
    This function plots the roc curve
    """
    
    train_prob = model.predict_proba(X_train)[:,1]
    val_prob = model.predict_proba(X_val)[:,1]
    plt.figure(figsize=(7,7))
    for data in [[y_train, train_prob],[y_val, val_prob]]: # ,[y_test, test_prob]
        fpr, tpr, threshold = roc_curve(data[0], data[1])
        plt.plot(fpr, tpr)
    annot(fpr, tpr, threshold)
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.ylabel('TPR (power)')
    plt.xlabel('FPR (alpha)')
    if test:
        plt.legend(['train','val'])
    else:
        plt.legend(['train', 'test'])
    plt.show()
```

```python
def annot_final(fpr,tpr,thr):
    """
    This function is called in the roc_plot function and serves to annotate the roc curves.
    """
    k=0
    for i,j in zip(fpr,tpr):
        if k %50 == 0:
            plt.annotate(round(thr[k],2),xy=(i,j), textcoords='data') 
        k+=1
```

```python
#Lets get the test data and plot it's curve on our favourite ROC plot. 
def roc_plot_final(model,X_train,y_train,X_val,y_val,X_test,y_test): #test values added to this new function!
    
    """
    This function plots the final roc curve with train, validation and test data now included
    """
    
    train_prob = model.predict_proba(X_train)[:,1]
    val_prob = model.predict_proba(X_val)[:,1]
    test_prob = model.predict_proba(X_test)[:,1] #This line added to roc_plot_final as we need test data now!
    plt.figure(figsize=(7,7))
    for data in [[y_train, train_prob],[y_val, val_prob],[y_test, test_prob]]:
        fpr, tpr, threshold = roc_curve(data[0], data[1])
        plt.plot(fpr, tpr)
    annot(fpr, tpr, threshold)
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.ylabel('TPR (power)')
    plt.xlabel('FPR (alpha)')
    plt.legend(['train','val', 'test'])
    plt.show()
```

## Next we'll define our false positives, true positives, false negatives and true negatives and assign a cost against each of these

As always when assigning cost it is difficult to compare tangible financial costs with more human percieved costs 
such as time, health, safety, political, environmental and education. Eventually these factors eventually boil down to some financial value but to assign the value requires crude estimates. We will also force the assumption that the cost is always born by the same stakeholder, in this case the Tasmanian government. 

```python

#FALSE POSITIVE
#A false positive in our case is when we predict a faulty pump but the pump is functioning fine in reality
#The costs hear could include sending an operative to investigate + administrative costs!
fpc = 500+250

#TRUE POSITIVE
#A true positive in our case is when we predict a faulty pump and the pump indeed is found to be faulty. 
#We must record a true cost here as the cost of repair. We need to estimate an average repair cost! and 
#an additional factor for the short term intangibles. + administrative costs. 
tpc = 3000+1000+500


#FALSE NEGATIVE
#A false negative in our case is when we predict a pump is functioning fine but in reality is actually faulty
#As this may go unnoticed for a while with no initial costs we need to ensure that the eventual cost of this repair 
#is not the average repair cost but the average maximum repair cost due to further deterioration over time. And also 
#for any intangibles extended over time. We should bear in mind this could include a completely new bore hole!!
#+ administrative costs
fnc = 10000+3000+1500


#TRUE NEGATIVE
#A true negative in our case is when we predict the pump to be functioning and it is also functioning in reality
tnc = 0 

print(f'Our false positive cost is {fpc} points, our true positive cost is {tpc} points, our false negative cost\
 is {fnc} points and our true negative cost is {tnc} points!')
```

#### We calculate our m parameter usng the metz threhsold equation 

Equation: m = (1-prev)/prev * (fpc-tnc)/(fnc-tpc)

```python
m = ((1-prev)/prev) * ((fpc-tnc)/(fnc-tpc))
m
```

#### We calculate our f-measure to locate our preferred threshold

Equation: fm = tpr-(m*fpr)

```python
fm_list = tpr_test-(m*fpr_test)
fm_threshold = list(zip(fm_list.tolist(), thresholds_test.tolist())) #Returns a list of tuples
fm_threshold[0:10]
```

```python
#Calculating our maximum fm value we can utilise the recommended threshold should be set at ~0.26!
max(fm_threshold,key=lambda item:item[0])
```

#### Now we'll produce a confusion matrix to illustrate the performance of our final model. 

```python
y_pred_rfc = final_model.predict(X_train_f)
```

```python
def confusion_matrix(y_true, y_pred):
    cm = {'TP': 0, 'TN': 0, 'FP': 0, 'FN': 0}
    
    for ind, label in enumerate(y_true.tolist()):
        pred = y_pred_rfc[ind]
        if label == 1:
            if label == pred:
                cm['TP'] += 1
            else:
                cm['FN'] += 1
    
        else:
            if label == pred:
                cm['TN'] += 1
            else:
                cm['FP'] += 1
                
    return cm


confusion_matrix(y_test, y_pred_rfc)
            
        
```

```python
def create_plot_of_feature_importances(model, X):
    ''' 
    This function allows us to view the relative importance of the selected features
    '''
    
    feat_importances = model.feature_importances_

    features_and_importances = zip(X.columns, feat_importances)
    features_and_importances = sorted(features_and_importances, 
                                     key = lambda x: x[1], reverse=True)
    
    features = [i[0] for i in features_and_importances]
    importances = [i[1] for i in features_and_importances]
    
    
    label = importances
#     for i in range(len(importances)):
#         label.append(round(i))
    
    plt.figure(figsize=(10, 15))
    plt.barh(features[0:15], importances[0:15]) #Plots the top 15 most important features!
    plt.gca().invert_yaxis()
    plt.title('Feature Importances', fontsize=15)
    plt.xlabel('Importance', fontsize=12)
    #print(label)
    for i in range(len(label)):
        plt.annotate(str(label[i]), xy=(label[i], features[i])) 
    
    #Check out if any of the categorical features have large importance when combined??
    print('management features account for {} importance ' .format(label[3]+label[7]))
    print('source_type features account for {} importance ' .format(label[5]+label[14]))
    print('area features account for {} importance' . format(label[6]+label[12]))

   
```

## Key Takeaways

Take Away 1
Of all the faulty pumps in the set, we correctly identified 71% of them!!

Take Away 2
Of all the pumps that we said were faulty only 60% were actually faulty which means that if you were to take up our model you could be send a lot of engineers on unnecessary trips to investigate.

Because of the large relative cost of false negatives in this case, and the fact that the nature of this project is likely to have a greater tolerance for false positives than false negatives, our threshold is set low. 

To put another way our client is likely to value reduction in the amount of faulty wells that we miss over the amount of functioning wells that we incorrectly predict are faulty. 

This situation could be improved if our model was able to separate the negative and positive distributions more effectively meaning the trade off between FP and FN could be minimised.
